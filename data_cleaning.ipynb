### Import Libraries

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

### Load and Clean College Player Dataset

```python
# load in the raw college player dataset
college_raw_df = pd.read_csv('college_player_data.csv', skiprows=1)
display(college_raw_df.head())
print(college_raw_df.shape)
```

```python
# Clean the college dataset
# drop columns that aren't needed
college_df = college_raw_df.drop(columns=['Draft Team', 'Rk', 'From', 'To', 'Team', '-9999'])
college_df = college_df.rename(columns={'Draft College':'College'})
display(college_df.head())
print(print(college_df.columns))
```

```python
college_df.isnull().sum()
```

```python
# save the dataset
college_df.to_csv('college_player_data_cleaned.csv')
```

### Load and Clean NBA Player Dataset

```python
nba_raw_df = pd.read_csv('nba_players.csv', skiprows=1)
display(nba_raw_df.head())
print(nba_raw_df.shape)
print(nba_raw_df.columns)
```

```python
# drop columns that aren't needed
nba_df = nba_raw_df.drop(columns=['Rk', 'Team', '-9999'])
```

```python
# add pick number to nba dataset
pick_df = college_df[['Player', 'Round', 'Pick']]
display(pick_df.head())
```

```python
nba_df = pd.merge(college_df, nba_df, on='Player', how='inner')
display(nba_df.head())
print(nba_df.shape)
```

```python
nba_df.isnull().sum()
```

```python
# save the dataset
nba_df.to_csv('nba_player_data_cleaned.csv')
```
